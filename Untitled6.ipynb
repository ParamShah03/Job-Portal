```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeClassifier
```

```python
train_path = '/content/gdrive/MyDrive/LOC/jobclassinfo2.csv'
train_df = pd.read_csv(train_path)
train_df
```

```python
train_df.columns
```

```python
label = LabelEncoder()

a = train_df['JobFamilyDescription'].to_numpy()
b = train_df['JobClassDescription'].to_numpy()
c = train_df['PG'].to_numpy()

train_df.drop(['JobFamilyDescription','JobClassDescription','PG','ID'],axis=1, inplace=True)

train_df['JobFamilyDescription'] = label.fit_transform(a)
train_df['JobClassDescription'] = label.fit_transform(b)
train_df['PG'] = label.fit_transform(c)
train_df
```

```python
X_train, X_test, y_train, y_test=train_test_split(train_df,train_df['JobFamilyDescription'],test_size=0.3,random_state=42)
```

```python
scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
y_train = y_train.to_numpy()
y_test = y_test.to_numpy()
```

```python
dct = DecisionTreeClassifier()
dct.fit(X_train,y_train)
print(dct.score(X_train,y_train))
print(dct.score(X_test,y_test))
```
